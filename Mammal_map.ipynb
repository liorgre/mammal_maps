```python
# %run -m biomass_data_analysis
import pandas as pd
data =pd.read_csv('Predicted_densities.csv')
data=data.drop([ 'estimated_mass','estimated_pop'],axis = 1)
data = data[data.binomial != 'Sus scrofa']
```

```python
import geopandas as gpd
df = gpd.read_file('TERRESTRIAL_MAMMALS/TERRESTRIAL_MAMMALS.shp')
df = df[df.category != 'EX']
df = df[df.binomial != 'Sus scrofa']
range_polygons = df.loc[(df['legend'] == 'Extant & Introduced (resident)') |
                        (df['legend'] == 'Extant & Origin Uncertain (resident)')|
                       (df['legend'] == 'Extant & Reintroduced (resident)')|
                       (df['legend'] == 'Extant & Vagrant (seasonality uncertain)')|
                       (df['legend'] == 'Extant (non breeding)')|
                       (df['legend'] == 'Extant (resident)')|
                       (df['legend'] == 'Probably Extant & Origin Uncertain (resident)')|
                       (df['legend'] == 'Probably Extant (resident)')|
                       (df['legend'] == 'Reintroduced')]
```

```python
continents = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
continents = continents[continents.continent!='Antarctica']
# continents = continents[['continent', 'geometry']]
# continents = continents.dissolve(by='continent')
# continents
continents_merc = continents.to_crs("EPSG:6933")
# continents_merc.plot()
# continents_merc
```

```python
continents = gpd.read_file('continent_shapefile/continent.shp')
continents = continents[continents.CONTINENT!='Antarctica']
continents_merc = continents.to_crs("EPSG:6933")
continents_merc.plot()
continents_merc.head()
```

```python
# continents_merc[continents_merc.CONTINENT=='Australia'].area*10**(-6)
```

```python
range_polygons_merc = range_polygons.to_crs("EPSG:6933")
```

```python
range_polygons_w_data = range_polygons_merc.merge(data, on ='binomial')
range_polygons_w_data
```

```python
data_by_cont = gpd.overlay(range_polygons_w_data, continents_merc, how='intersection')
```

```python
# continents_merc
```

```python
data_by_cont = data_by_cont.assign(area_km_2 = data_by_cont.geometry.area*10**(-6))
data_by_cont = data_by_cont.assign(population = data_by_cont.pop_density*data_by_cont.area_km_2)
data_by_cont = data_by_cont.assign(total_mass = data_by_cont.population*data_by_cont.AdultBodyMassG)
```

```python
data_by_cont = data_by_cont[data_by_cont.binomial != 'Sus scrofa']
# data_by_cont
```

```python
data_by_cont.to_file("data_by_country.shp")
```

```python
data_by_cont_grouped = data_by_cont.groupby(['name']).sum()
data_by_cont_grouped
```

```python
data_by_cont_grouped = data_by_cont_grouped.merge(continents_merc,on ='name')
data_by_cont_grouped = GeoDataFrame(data_by_cont_grouped)
data_by_cont_grouped = data_by_cont_grouped.assign(total_mass_Mt=data_by_cont_grouped.total_mass*10**(-12))
data_by_cont_grouped = data_by_cont_grouped.assign(norm_mass=data_by_cont_grouped.total_mass/data_by_cont_grouped.geometry.area)

data_by_cont_grouped
```

```python
import matplotlib.pyplot as plt
data_by_cont_grouped = GeoDataFrame(data_by_cont_grouped)
# data_by_cont_grouped.plot(column = 'total_mass_Mt')
plt.rcParams["figure.figsize"] = (30,10)
font = {'family' : 'normal',
        'weight':'normal',
        'size'   : 22}

plt.rc('font', **font)

fig, ax = plt.subplots(1, 1)
plt.title('Sum of mammal mass [Mt] per country ')
plt.axis('off')

data_by_cont_grouped.plot(column = 'total_mass_Mt', ax=ax, legend=True)

```
